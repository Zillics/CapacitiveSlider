```python
import pandas as pd
df = pd.read_csv('sliderData/rawData2.csv',header=None,names=['v1','v2'])
```

```python
df.plot(y=['v1','v2'])
df['diff'] = df['v1']-df['v2']
df.plot(y=['diff','v1','v2'])
```

```python
import numpy as np
M = 20
mean_range = (np.arange(1000/M)*M).astype(int)
df['mean'] = 0
for i in mean_range:
    v1_mean = df['v1'][i:(i+M)].values.mean()
    v2_mean = df['v2'][i:(i+M)].values.mean()
    df.loc[i:(i+M),'mean'] = v1_mean-v2_mean
```

```python
df[0:1000].plot(y=['diff','mean'])
```

```python
M=20
mean_range = (np.arange(1000/M)*M).astype(int)
df['mean_diff'] = 0
df2 = df[0:1000].copy()
for i in mean_range:
    mean_i = df['diff'][i:(i+M)].values.mean()
    #print(mean_i)
    df2.loc[i:(i+M),'mean_diff'] = mean_i
```

```python
df2.plot(y=['mean'])
```
